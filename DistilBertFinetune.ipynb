```python
from datasets import load_dataset
dataset = load_dataset("go_emotions")
print("Dataset sample:", dataset["train"][0])
label_names = dataset["train"].features["labels"].feature.names
print(label_names)
```

Fetches a dataset of Reddit comments labeled with emotions like "joy," "anger," etc.


```python
import pandas as pd
df = pd.DataFrame(dataset["train"])
mood_map = {
    0: "happy",   # admiration → happy
    8: "happy",   # desire → happy
    23: "happy",  # relief → happy

    # Keep other mappings
    1: "happy",   # amusement
    4: "happy",   # approval
    5: "happy",   # caring
    13: "happy",  # excitement
    15: "happy",  # gratitude
    17: "happy",  # joy
    18: "happy",  # love
    20: "happy",  # optimism
    21: "happy",  # pride

    9: "sad",     # disappointment
    16: "sad",    # grief
    24: "sad",    # remorse
    25: "sad",    # sadness

    2: "angry",   # anger
    3: "angry",   # annoyance
    10: "angry",  # disapproval
    11: "angry",  # disgust

    6: "fearful",  # confusion
    12: "fearful", # embarrassment
    14: "fearful", # fear
    19: "fearful", # nervousness

    22: "surprised",  # realization
    26: "surprised",  # surprise

    # Reduced neutral mappings
    7: "neutral",   # curiosity
    27: "neutral"   # explicit neutral
}

def map_mood(labels):
    """Map labels to a single mood, defaulting to 'neutral' if no match."""
    for label in labels:
        if label in mood_map:
            return mood_map[label]
    return "neutral"

df["mood"] = df["labels"].apply(map_mood)

df = df[["text", "mood"]].sample(n=10000, random_state=42)
print("Mood distribution:\n", df["mood"].value_counts())
```

Groups complex emotions into 6 basic moods. Example: "joy" becomes "happy"

```python
from transformers import DistilBertTokenizer
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
text = "I’m feeling great!"
tokens = tokenizer(text, padding="max_length", truncation=True,max_length=128,return_tensors="pt")
print(tokens)
```

Converts sentences into tokens that DistilBERT understands.
Example: "great" → [2307].

```python
import torch
from torch import nn
from transformers import DistilBertForSequenceClassification, Trainer, TrainingArguments

MOOD_LABELS = ["happy", "sad", "angry", "fearful", "surprised", "neutral"]

model = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=len(MOOD_LABELS), #moods
    id2label={i: label for i, label in enumerate(MOOD_LABELS)},
     label2id={label: i for i, label in enumerate(MOOD_LABELS)}
)

for param in model.distilbert.parameters():
    param.requires_grad = True



train_df = df.sample(frac=0.8, random_state=42)
val_df = df.drop(train_df.index)


train_encodings = tokenizer(train_df["text"].tolist(), truncation=True, padding=True, max_length=128)
val_encodings = tokenizer(val_df["text"].tolist(), truncation=True, padding=True, max_length=128)

train_labels = [model.config.label2id[mood] for mood in train_df["mood"]]
val_labels = [model.config.label2id[mood] for mood in val_df["mood"]]

class MoodDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = MoodDataset(train_encodings, train_labels)
val_dataset = MoodDataset(val_encodings, val_labels)
```

```python
from torch import nn
from transformers import Trainer

class_weights = torch.tensor(
    [1.0,  # happy
     10.0,  # sad
     5.0,  # angry
     8.0,  # fearful
     12.0,  # surprised
     0.5   # neutral
], dtype=torch.float32).to("cpu")

class CustomTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False,**kwargs):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        loss_fct = nn.CrossEntropyLoss(weight=class_weights)
        loss = loss_fct(outputs.logits, labels)
        return (loss, outputs) if return_outputs else loss
```

```python
from sklearn.metrics import classification_report

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    return classification_report(labels, preds, target_names=MOOD_LABELS, output_dict=True)
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=5e-5,  # Increased from 3e-5
    per_device_train_batch_size=32,
    num_train_epochs=5,
    weight_decay=0.01,
    fp16=False,
    gradient_accumulation_steps=1,
    warmup_ratio=0.2,
    gradient_checkpointing=True,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    logging_steps=50,
    warmup_steps=100,
    lr_scheduler_type="cosine",
)


trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics
)

# Add EarlyStoppingCallback
from transformers import EarlyStoppingCallback
trainer.add_callback(EarlyStoppingCallback(
    early_stopping_patience=3,  # Stop if no improvement for 2 epochs
    early_stopping_threshold=0.01
))

trainer.train()
```

DistilBertForSequenceClassification: Loads DistilBERT and adds a classification layer for moods.

Trainer: Handles the training loop (forward pass, loss calculation, backpropagation).

TrainingArguments: Sets hyperparameters like learning rate and batch size.
